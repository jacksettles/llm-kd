```python
from data import Dataset
import numpy as np
from tqdm.notebook import tqdm
```

```python
data = Dataset("babylm_data/tokenized/babylm_final_dataset_16-train.pkl")
```

```python
len(data)
```

```python
batchsize_count = {}
sentsize_count = {}
total_tokens_16 = 0
total_tokens = 0
for i in tqdm(range(len(data)), total=454236):
    batch = data[i][0]
    batch_size = batch.size(0)
    num_tokens = batch.size(1)
    if batch_size not in batchsize_count.keys():
        batchsize_count[batch_size] = 1
    else:
        batchsize_count[batch_size] += 1
        
    if num_tokens not in sentsize_count.keys():
        sentsize_count[num_tokens] = 1
    else:
        sentsize_count[num_tokens] += 1
        
    if batch_size == 16:
        total_tokens_16 += (16*num_tokens)
    else:
        
        print(f"Batch size not 16 - batch with sentence length {num_tokens} will not be in dataset")
    
    total_tokens += (batch_size*num_tokens)
    
print(f"Number of tokens in the batch size of 16 batches only: {total_tokens_16}")
print(f"Number of total tokens: {total_tokens}")
print(f"Percentage of total from batch size 16: {(total_tokens_16/total_tokens)*100:.2f}%")
```
